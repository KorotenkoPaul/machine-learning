```python
%config InlineBackend.figure_format = 'svg'
plt.style.use('ggplot')
```

```python
# Data
x = [x for x in range(1, 8)]
y = [3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]
```

```python
ax = plt.subplot()

ax.yaxis.set_major_locator(MultipleLocator(base=1))
ax.plot(x, y);
```

```python
ax = plt.subplot()

ax.yaxis.set_major_locator(MultipleLocator(base=1))
ax.scatter(x, y);
```

#### task 2

```python
# Data
t = np.linspace(0, 10, 51)
f = np.cos(t)
```

```python
ax = plt.subplot()

ax.yaxis.set_major_locator(MultipleLocator(base=1))
ax.set_title(label=' f(t)')
ax.set_xlabel(xlabel=' t')
ax.set_xlim(left=0.5, right=9.5)
ax.set_ylabel(ylabel=' f')
ax.set_ylim(bottom=-2.5, top=2.5)
ax.plot(t, f, c='g');
```

#### task 3 *

```python
# Data
x = np.linspace(-3, 3, 51)
y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

for ax, y, num in zip(axs.flatten(), [y1, y2, y3, y4], [1, 2, 3, 4]):
    ax.plot(x, y)
    ax.set_title(label=f' y{num}')

# Should have just used constrained_layout=True in Figure object
plt.subplots_adjust(hspace=0.3)
plt.subplots_adjust(wspace=0.3)

axs[0, 0].set_xlim(left=-5, right=5);
```
